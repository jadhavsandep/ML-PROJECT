 ##import libarires

```python
# Cell 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for better visualization aesthetics
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

 ##Load data and preview

```python
# Cell 2: Load Data and Preview
data = pd.read_csv('mental_health_diagnosis_treatment_.csv')
data.head()
```

#Data cleaning

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

# Distribution of Age

```python
# Cell 5: Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#Diagnosis Distribution by Gender


```python
# Cell 6: Diagnosis Distribution by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Diagnosis', hue='Gender')
plt.title('Diagnosis Distribution by Gender')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

#Treatment Outcome Distribution

```python
# Cell 7: Treatment Outcome Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Outcome')
plt.title('Treatment Outcome Distribution')
plt.xlabel('Treatment Outcome')
plt.ylabel('Count')
plt.show()
```

#Correlation Heatmap

```python
# Cell 8: Correlation Heatmap (Numeric Columns Only)
# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

 #Distribution of Symptom Severity

```python
# Histogram for Symptom Severity
plt.figure(figsize=(8, 5))
sns.histplot(data['Symptom Severity (1-10)'], bins=10, kde=True)
plt.title('Distribution of Symptom Severity')
plt.xlabel('Symptom Severity (1-10)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot for Symptom Severity
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['Symptom Severity (1-10)'])
plt.title('Box Plot of Symptom Severity')
plt.xlabel('Symptom Severity (1-10)')
plt.show()
```

 #Symptom Severity by Treatment Outcome

```python
# Cell 10: Symptom Severity by Treatment Outcome
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Outcome', y='Symptom Severity (1-10)')
plt.title('Symptom Severity by Treatment Outcome')
plt.xlabel('Treatment Outcome')
plt.ylabel('Symptom Severity (1-10)')
plt.show()
```

 #Scatter Plot of Symptom Severity vs. Mood Score and Sleep Quality

```python
# Cell 11: Scatter Plot of Symptom Severity vs. Mood Score and Sleep Quality

# Scatter plot for Symptom Severity vs. Mood Score
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Symptom Severity (1-10)', y='Mood Score (1-10)')
plt.title('Symptom Severity vs. Mood Score')
plt.xlabel('Symptom Severity (1-10)')
plt.ylabel('Mood Score (1-10)')
plt.show()

# Scatter plot for Symptom Severity vs. Sleep Quality
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Symptom Severity (1-10)', y='Sleep Quality (1-10)')
plt.title('Symptom Severity vs. Sleep Quality')
plt.xlabel('Symptom Severity (1-10)')
plt.ylabel('Sleep Quality (1-10)')
plt.show()
```

#Symptom Severity vs. Adherence to Treatment

```python
# Cell 12: Symptom Severity vs. Adherence to Treatment
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Symptom Severity (1-10)', y='Adherence to Treatment (%)')
plt.title('Symptom Severity vs. Adherence to Treatment')
plt.xlabel('Symptom Severity (1-10)')
plt.ylabel('Adherence to Treatment (%)')
plt.show()
```

#Treatment Progress by Therapy Type

```python
# Cell 13: Treatment Progress by Therapy Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Therapy Type', y='Treatment Progress (1-10)')
plt.title('Treatment Progress by Therapy Type')
plt.xlabel('Therapy Type')
plt.ylabel('Treatment Progress (1-10)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cell 13: Treatment Progress by Therapy Type (with Different Colors)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Therapy Type', y='Treatment Progress (1-10)', palette="Set2")
plt.title('Treatment Progress by Therapy Type')
plt.xlabel('Therapy Type')
plt.ylabel('Treatment Progress (1-10)')
plt.xticks(rotation=45)
plt.show()
```

 #Mood Score Over Time

```python
# Cell 14: Mood Score Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Treatment Start Date', y='Mood Score (1-10)')
plt.title('Mood Score Trend Over Time')
plt.xlabel('Treatment Start Date')
plt.ylabel('Mood Score (1-10)')
plt.xticks(rotation=45)
plt.show()
```

#Mood Score by Treatment Duration

```python
# Cell 15: Mood Score by Treatment Duration
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Treatment Duration (weeks)', y='Mood Score (1-10)', hue='Therapy Type', palette="Set1")
plt.title('Mood Score by Treatment Duration')
plt.xlabel('Treatment Duration (weeks)')
plt.ylabel('Mood Score (1-10)')
plt.legend(title='Therapy Type')
plt.show()
```
